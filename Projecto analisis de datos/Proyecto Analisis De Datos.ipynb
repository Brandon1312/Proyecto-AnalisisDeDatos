```python
df_BienesRaices.rename({'Serial Number': 'serialNumber'},axis=1, inplace=True)
df_BienesRaices.rename({'List Year': 'listYear'},axis=1, inplace=True)
df_BienesRaices.rename({'Date Recorded': 'dateRecorded'},axis=1, inplace=True)
df_BienesRaices.rename({'Town': 'town'},axis=1, inplace=True)
df_BienesRaices.rename({'Assessed Value': 'assessedValue'},axis=1, inplace=True)
df_BienesRaices.rename({'Sale Amount': 'saleAmount'},axis=1, inplace=True)
df_BienesRaices.rename({'Sales Ratio': 'salesRatio'},axis=1, inplace=True)
df_BienesRaices.rename({'Property Type': 'propertyType'},axis=1, inplace=True)
df_BienesRaices.rename({'Residential Type': 'residentialType'},axis=1, inplace=True)
df_BienesRaices.rename({'Non Use Code': 'nonUseCode'},axis=1, inplace=True)
df_BienesRaices.rename({'Assessor Remarks': 'assessorRemarks'},axis=1, inplace=True)
df_BienesRaices.rename({'OPM remarks': 'oPMremarks'},axis=1, inplace=True)
df_BienesRaices.rename({'Location;;;;;': 'location'},axis=1, inplace=True)
df_BienesRaices.columns
```

```python
gf = df_BienesRaices.groupby(['listYear']).count()
gf = gf.iloc[:,:1]
gf = gf.sort_values('serialNumber', ascending=False)
gf
```

```python
ingresos_anuales = df_BienesRaices.groupby("listYear")["saleAmount"].sum()
fig, ax = plt.subplots(1,1, figsize =(15,6.9), dpi=100)


ingresos_anuales.plot(kind='bar', color='skyblue')
plt.title('Ingreso Anual de Propiedades Vendidas',horizontalalignment = 'left',x = 0.001,
             y = 0.94,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Liberation Sans Narrow")
plt.xlabel('A√±o')
plt.ylabel('Ingreso Anual ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
ingresos_anuales = df_BienesRaices.groupby("listYear")["saleAmount"].sum()


fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, sharey=True, figsize=(15, 6.9), dpi=100)

pricemin = ingresos_anuales.min()


ax1.plot(ingresos_anuales.index, ingresos_anuales, lw=2)
ax1.set_ylabel('Ingreso Anual ($)')


ax2.fill_between(ingresos_anuales.index, pricemin, ingresos_anuales, alpha=0.7)
ax2.set_ylabel('Ingreso Anual ($)')

# Configuraci√≥n del t√≠tulo
title_props = {'horizontalalignment': 'left',
               'x': 0.001,
               'y': 0.98,
               'transform': fig.transFigure,
               'fontsize': 20,
               'fontweight': 'bold',
               'fontfamily': 'serif',
               'fontname': 'Liberation Sans Narrow'}
fig.suptitle('Ingreso Anual de Propiedades Vendidas', **title_props)

# Etiquetas de ejes y t√≠tulo
ax1.set_xlabel('A√±o')
ax2.set_xlabel('A√±o')
plt.xticks(rotation=45)

# Ajuste de dise√±o y mostrar la gr√°fica
fig.autofmt_xdate()
plt.tight_layout()
plt.show()
```

<h4 Style='color:#009989'> La respuesta sobre la pregunta de cual es el costo promedio de las propiedades comerciales y residenciales, grafica con barras </h4>

```python

tipos_deseados = ["Residential", "Commercial"]
df_filtrado = df_BienesRaices[df_BienesRaices['propertyType'].isin(tipos_deseados)]
costo_promedio_por_tipo = df_filtrado.groupby('propertyType')['saleAmount'].mean()
circulo= plt.Circle((0,0), 0.7 , color='white')
plt.figure(figsize=( 15, 6),dpi=80, )
plt.pie(costo_promedio_por_tipo, labels=costo_promedio_por_tipo.index , autopct='%1.1f%%', startangle=25) 
plt.title('Distribuci√≥n del Costo Promedio por Tipo de Propiedad')
plt.axis('equal')  
plt.legend()
plt.show()
```

```python
# Calcular el margen de ganancia por propiedad
df_BienesRaices['profitMargin'] = ((df_BienesRaices['saleAmount'] - df_BienesRaices['assessedValue']) / df_BienesRaices['assessedValue']) * 100

# Calcular el margen promedio por tipo de propiedad
margen_promedio_por_tipo = df_BienesRaices.groupby('propertyType')['profitMargin'].mean()

print(margen_promedio_por_tipo)
```
